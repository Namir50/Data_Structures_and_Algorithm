```python
def fact_iter(n):
    answer = 1  #Here it is 1 opertion
    while n >1:  #Here it is 1 operation
        answer *= n  #Here are 2 operation
        n-=1 #Here are 2 operation
    return answer 
```

So total operations are 6 with input n, so 1 additive operation and 5 multiplicative operations<br>
The equation would be 1 +5n<br>
Removing constants we will get n so the time complexity is O(n)<br>
Hence we get a linear relationship between T(Time) and n(Input)

If the equation is n^2 + 2n + 2 <br>
we can say there are 2 nested loops(2n), and 1 loop with 2 operations and 2 operations outside the loops<br>
Removing additive and multiplicative constants we get n^2 + n, we have to consider biggest factor<br>
Hence time complexity is O(n^2)<br>
Hence the relationship between time and input is quadratic

If the equation is log(n) + n + 4<br>
Removing contants we get log(n) + n<br>
Considering biggest factors, the time complexity is 0(n)<br>
Hence linear relationship

if the equation is 0.001*n*log(n) + 300n<br>
Removing contants we get n*log(n) + n<br>
Considering biggest factors, the time complexity is O(n*log(n))

If the equation is 2n^30 + 3^n<br>
The time complexity will be O(3^n)
